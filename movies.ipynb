```python
df = df[['title', 'overview', 'genres', 'tagline', 'vote_average', 'popularity']]
df
```

```python
df = df.dropna(subset=['title'])
df['overview'] = df['overview'].fillna('')
df.isnull().sum()
```

```python
import ast 
df['genres'] = df['genres'].apply(
    lambda x: " ".join([i['name'] for i in ast.literal_eval(x)])
)
df.head()
```

```python
df['tagline'] = df['tagline'].fillna('')
df.isnull().sum()
```

```python
df['tags'] = df['overview'] + ' ' + df['genres'] + ' ' + df['tagline']
print(df['tags'][0])
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# 3. Download data
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stopword = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def  preprocess_text(text):
    # lower case
    text = str(text.lower())

    # remove the punctuations
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    words = text.split()

    # remove stopwords
    words = [word for word in words if word not in stopword]

    # lemmatize
    words = [lemmatizer.lemmatize(word) for word in words]


    return " ".join(words)

df['tags'] = df['tags'].apply(preprocess_text)
df['tags']
```

```python
df = df.reset_index(
    drop=True
)

indices = pd.Series(
    df.index,
    index=df.title
).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', max_features=50000, ngram_range=(1, 2))
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(title, n=10):
    # Check if movie exists
    if title not in indices:
        return "Movie not found"
    
    # Get the index of the movie
    idx = indices[title]

    # Calculate similarity for ONLY this movie against all others
    # This is RAM-safe! It only computes 1 row vs 45,000 rows.
    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    # Sort the movies based on similarity scores (descending order)
    # We use argsort() to get the indices of the sorted scores
    similar_idx = sim_score.argsort()[::-1]

    # Get the top n (skipping the 0th one, which is the movie itself)
    similar_idx = similar_idx[1:n+1]

    # Return the titles
    return df['title'].iloc[similar_idx]
recommend('Toy Story')
```

```python
recommend('Avatar', 5)
```

```python
import pickle
pickle.dump(
    tfidf_matrix, 
    open('tfidf_matrix.pkl', 'wb')
)

pickle.dump(
    indices,
    open('indices.pkl', 'wb')
)

df.to_pickle(
    'df.pkl'
)

pickle.dump(
    tfidf,
    open('tfidf.pkl', 'wb')
)

print('successfully save all file👌')
```
